```python
import numpy as np
import math
from math import sqrt
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import execute, Aer, BasicAer
from qiskit.tools.visualization import circuit_drawer
from qiskit.tools.visualization import plot_histogram
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
from qiskit import transpile
```

```python
qc = QuantumCircuit(3)        # Buat sirkuit kuantum dengan 4 qubit: q2, q1, q0
qc.h(0)
qc.h(1)
qc.h(2)
qc.barrier()  
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc
psi_a=Statevector(qc)
psi_a.draw('latex')
```

```python
# kedua
qc.x(0)
qc.cx(1,2)
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc 2
psi_b=Statevector(qc)
psi_b.draw('latex')
```

```python
# ketiga
qc.y(2)
qc.barrier()  
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc 3
psi_c=Statevector(qc)
psi_c.draw('latex')
```

```python
# keempat
qc.ccx(0,1,2)
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc 4
psi_d=Statevector(qc)
psi_d.draw('latex')
```

```python
# kelima
qc.x(0)
qc.y(1)
qc.barrier()  
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc 5
psi_e=Statevector(qc)
psi_e.draw('latex')
```

```python
# keenam
qc.z(1)
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc 6
psi_f=Statevector(qc)
psi_f.draw('latex')
```

```python
# ketuju
qc.swap(0,2)
qc.barrier() 
qc.draw('mpl')
```

```python
# Keadaan Akhir dari qc 7
psi_f=Statevector(qc)
psi_f.draw('latex')
```

```python
qc. measure_all()             # Pengukuran pada seluruh qubit
qc.draw('mpl')                # Tampilkan gambar sirkuit di layar
```
